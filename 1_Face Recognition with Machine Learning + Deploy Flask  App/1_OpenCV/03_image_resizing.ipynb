# Resize

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
#loading image
img = cv2.imread('./data/flying_bird.jpg')

print(img.shape)
plt.imshow(img)
plt.show()

```

###  Task
1. Shink the image
2. Enlarge the image

```python
# shinking image
# width = columns , height = rows
img_shink = cv2.resize(img,dsize=(320,240),interpolation=cv2.INTER_AREA)

# enlarge 
# cv2.INTER_CUBIC, cv2.INTER_LINEAR
img_zoom = cv2.resize(img,dsize=(1280,960),interpolation=cv2.INTER_CUBIC) # 
```

```python
img_shink.shape, img_zoom.shape
```

```python
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.imshow(img_shink)
# plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(img_zoom)
# plt.axis('off')
plt.show()
```
