```python
import numpy as np
import cv2
import matplotlib.image as matimg
from PIL import Image

import matplotlib.pyplot as plt
```

```python
# OpenCV - BGR - returns numpy array
img_cv = cv2.imread('./data/flying_bird.jpg')
# Matplotlib - RGB - returns numpy array
img_mat = matimg.imread('./data/flying_bird.jpg')
# PIL - RGB - returns object
img_pil = Image.open('./data/flying_bird.jpg')
```

```python
# OpenCV - BGR
print(img_cv)
```

```python
# matplotlib -RGB
print(img_mat)
```

```python
img_pil
```

```python
# converting PIL object to array
img_pil_arr = np.array(img_pil)
```

```python
img_pil_arr
```

### Display Image

```python
# display image using opencv
cv2.imshow('OpenCV BGR',img_cv)
cv2.imshow('Matplotlib RGB',img_mat)
cv2.imshow('Pillo RGB',img_pil_arr)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# display image using matplotlib
plt.figure(figsize=(15,8))
plt.subplot(1,3,1)
plt.imshow(img_cv)
plt.axis('off')
plt.title('OpenCV BGR')

plt.subplot(1,3,2)
plt.imshow(img_mat)
plt.axis('off')
plt.title('Matplotlib RGB')

plt.subplot(1,3,3)
plt.imshow(img_pil_arr)
plt.axis('off')
plt.title('Pillow RGB')

plt.show()
```

### COLOR Space

- Color Conversion
- Gray Scale

```python
img_cv.shape
```

```python
# numpy slicing
# BGR
b = img_cv[:,:,0]
g = img_cv[:,:,1]
r = img_cv[:,:,2]
```

```python
plt.figure(figsize=(15,8))
plt.subplot(1,3,1)
plt.imshow(b,cmap='gray')
plt.axis('off')
plt.title('Blue Channel')

plt.subplot(1,3,2)
plt.imshow(g,cmap='gray')
plt.axis('off')
plt.title('Green Channel')

plt.subplot(1,3,3)
plt.imshow(r,cmap='gray')
plt.axis('off')
plt.title('Red Channel')

plt.show()
```

```python
# opencv split channels
b,g,r  = cv2.split(img_cv)
```

```python
plt.figure(figsize=(15,8))
plt.subplot(1,3,1)
plt.imshow(b,cmap='gray')
plt.axis('off')
plt.title('Blue Channel')

plt.subplot(1,3,2)
plt.imshow(g,cmap='gray')
plt.axis('off')
plt.title('Green Channel')

plt.subplot(1,3,3)
plt.imshow(r,cmap='gray')
plt.axis('off')
plt.title('Red Channel')

plt.show()
```

```python
b.shape, g.shape, r.shape, img_cv.shape
```

```python
# converting color
# bgr to rgb
img_cv_rgb = cv2.cvtColor(img_cv,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_cv_rgb)
plt.axis('off')
plt.show()
```

### GrayScale

```python
img_gray = cv2.cvtColor(img_cv,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img_gray,cmap='gray')
plt.axis('off')
plt.show()
```
